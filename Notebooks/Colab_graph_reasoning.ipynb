```python
# In[ ]:
# ---- Query set (domain questions) ----
Q = [
    "What are the knobs that can change the uniformity in radical si-etching process?",
    "How to increase the selectivity ratio of Si to oxide in ICP etching?",
    "How to reduce the particle in the dechuck step of etching?",
    "How to improve the cleaning or etching ability of Al particles?",
    (
        "Two-stage etching on poly-silicon:\n"
        "Stage-1: HAR etch, 30 nm width, 100 nm depth.\n"
        "Stage-2: selectivity 22:1 (poly-silicon:SiO2), extra 45 nm depth, base STI OX."
    ),
    "How to achieve ~70° taper using Cl2/HBr/O2 on silicon?",
    "On a trench 15 nm (W) × 75 nm (D), how to etch SiGe bottom 10 nm and SiP 5 nm?",
    "How to increase oxidizing power of O2/H2 radical on W sidewall in HAR (15×75 nm)?",
    "How to reduce SiN etching at the bottom of a 15×70 nm trench without bowing?",
    "How to reduce iso-dense CD loading on silicon etching?",
]
```

```python
print(Q[4])
```

```python
from transformers import pipeline

generator = pipeline("text-generation", model="mkychsu/tsmc_graph_preflexor_grpo", device="cuda")

```

```python
question = Q[4]
output = generator([{"role": "user", "content": question}], max_new_tokens=8192, return_full_text=False)[0]
print(output["generated_text"])
```
