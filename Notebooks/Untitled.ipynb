```python
import os
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel

os.environ["HF_HUB_OFFLINE"] = "1"

BASE_MODEL_DIR = "/home/mkychsu/pool/llm/Llama-3.1-8B-Instruct"
ADAPTER_DIR = "/home/mkychsu/workspace/LLM/GraphReasoning_working/Notebooks/TSMC/graph-preflexor-grpo/orpo_graph_model"  

device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.bfloat16 if device == "cuda" else torch.float32

assert os.path.isdir(BASE_MODEL_DIR), BASE_MODEL_DIR
assert os.path.isdir(ADAPTER_DIR), ADAPTER_DIR

print("BASE files:", os.listdir(BASE_MODEL_DIR))
print("ADAPTER files:", os.listdir(ADAPTER_DIR))

# tokenizer: load from adapter dir only if tokenizer files were saved there
tokenizer_dir = ADAPTER_DIR if os.path.exists(os.path.join(ADAPTER_DIR, "tokenizer_config.json")) else BASE_MODEL_DIR

tokenizer = AutoTokenizer.from_pretrained(
    tokenizer_dir,
    local_files_only=True,
    use_fast=False,
)

if tokenizer.pad_token_id is None:
    tokenizer.pad_token = tokenizer.eos_token

base_model = AutoModelForCausalLM.from_pretrained(
    BASE_MODEL_DIR,
    local_files_only=True,
    trust_remote_code=True,
    torch_dtype=dtype,
    device_map="auto" if device == "cuda" else None,
)

base_model.resize_token_embeddings(len(tokenizer))

model = PeftModel.from_pretrained(
    base_model,
    ADAPTER_DIR,
    local_files_only=True,
)

model.eval()
```
